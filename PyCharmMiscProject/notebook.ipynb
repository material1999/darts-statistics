```python
input_file_path = "/Users/matevass/Desktop/results.txt"

players = 12
```

```python
with open(input_file_path, "r") as input_file:
    lines = input_file.readlines()

num_matches = int(players * (players - 1) / 2)

for i in range(0, num_matches * 9, 9):
    group = lines[i:i+4]
    player_1 = str(group[0].strip())
    player_2 = str(group[1].strip())
    score_1 = int(group[2].strip())
    score_2 = int(group[3].strip())
    if score_1 > score_2:
        score_1 = str(score_1) + "*"
        score_2 = str(score_2)
    else:
        score_1 = str(score_1)
        score_2 = str(score_2) + "*"
    print(player_1 + "\t" + score_1 + "\t" + player_2 + "\t" + score_2)

# print()

semi_1 = lines[num_matches * 9 + 3:num_matches * 9 + 7]
player_1 = str(semi_1[0].strip())
player_2 = str(semi_1[1].strip())
score_1 = int(semi_1[2].strip())
score_2 = int(semi_1[3].strip())
if score_1 > score_2:
    score_1 = str(score_1) + "*"
    score_2 = str(score_2)
else:
    score_1 = str(score_1)
    score_2 = str(score_2) + "*"
print(player_1 + "\t" + score_1 + "\t" + player_2 + "\t" + score_2)

semi_2 = lines[num_matches * 9 + 11:num_matches * 9 + 15]
player_1 = str(semi_2[0].strip())
player_2 = str(semi_2[1].strip())
score_1 = int(semi_2[2].strip())
score_2 = int(semi_2[3].strip())
if score_1 > score_2:
    score_1 = str(score_1) + "*"
    score_2 = str(score_2)
else:
    score_1 = str(score_1)
    score_2 = str(score_2) + "*"
print(player_1 + "\t" + score_1 + "\t" + player_2 + "\t" + score_2)

# print()

final = lines[num_matches * 9 + 19:num_matches * 9 + 23]
player_1 = str(final[0].strip())
player_2 = str(final[1].strip())
score_1 = int(final[2].strip())
score_2 = int(final[3].strip())
if score_1 > score_2:
    score_1 = str(score_1) + "*"
    score_2 = str(score_2)
else:
    score_1 = str(score_1)
    score_2 = str(score_2) + "*"
print(player_1 + "\t" + score_1 + "\t" + player_2 + "\t" + score_2)
```
